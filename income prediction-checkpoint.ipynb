```python
import pandas as pd
import numpy as np
df = pd.read_csv(r'C:\JOEL\important-PDFs\WebDesign_Django_&_AI_Prateek_Joshi\Packt.Hands-On.Data.Science.and.Python.Machine.Learning\Hands-On-Data-Science-and-Python-Machine-Learning-master\datasets\adult.csv')
print (df.shape)
df.head()
```

```python
# though 'hours-per-week is limited, would still like to include it as a continuous variable 
cnames=['age','fnlwgt','hours-per-week','capital-gain','capital-loss']
# educational-num is numeric representation of variable 'education', so droping the variable 'education'
df = df.drop(axis=1,labels=['education'])
cat_names = ['workclass','educational-num','marital-status','occupation','relationship','race','gender','native-country','income']
print (df.columns)
```

```python
# missing values given by '?' have to replaced with 'np.nan' null values
df.replace('?',np.nan,inplace=True)

#creating missing dataframe function
def missing(df):
    miss = pd.DataFrame(df.isnull().sum())
    miss = miss.reset_index()
    miss = miss.rename(columns={'index':'variables',0:'percentage'})
    miss['percentage'] = (miss['percentage']/df.shape[0])*100
    miss = miss.sort_values(by='percentage',ascending=False)
    return miss

missing(df)
```

```python
# converting the categorical variable data to object type with numeric categrical-codes
for c in cat_names:
    df[c] = pd.Categorical(df[c])
    df[c] = df[c].cat.codes
    df[c] = df[c].astype('object')
    print (c,': ',df[c].unique())
    print (df[c].dtype,'\n')
    


from missingpy import KNNImputer # imputer library for python 3.5.2
# Doing KNN-imputation over the dataframe
df = pd.DataFrame(KNNImputer(n_neighbors=2, weights="uniform").fit_transform(df), columns = df.columns)

missing(df)
```

```python
# Let's check for outliers and remove them if present
for c in cnames:
    print ('outlier removal for: ',c)
    q75,q25 = np.percentile(df[c],[75,25])
    iqr = q75-q25
    maximum = q75+(iqr*1.5)
    minimum = q25-(iqr*1.5)
    df.loc[df[c]>maximum,c]=np.nan
    df.loc[df[c]<minimum,c]=np.nan
    
missing(df) # check missing values after outlier removal
# total percentage of missing value of any variable is below 30% so we'll impute them
```

```python
for i in ['hours-per-week','capital-gain','capital-loss','fnlwgt','age']: # list of missing value variables
    df[i] = df[i].fillna(df[i].median())
    
missing(df)    
```

```python
# all variables seem normally distributed, excepted capital-gain and capital-loss indicating only zero-values

df['capital-gain'].unique() #array([0.])
(df['capital-loss'].unique()) #array([0.])

# like seen visually,they have just 0 values and nothing more to contribute,let's remove them from our dataframe
df.drop(labels=['capital-gain','capital-loss'],axis=1,inplace=True)
cnames = ['age','fnlwgt','hours-per-week']

# Let's also normalize the continuous data (between 0 and 1)
for i in cnames:
    df[i] = (df[i]-min(df[i]))/(max(df[i])-min(df[i]))
    
df[cnames]    
```

```python
corr = df[cnames].corr()
#Set the width and hieght of the plot

f, ax = plt.subplots(figsize=(6,6))
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool), 
            cmap=sns.diverging_palette(220, 50, as_cmap=True),
            square=True, ax=ax,annot = True)
plt.show()
# no correlation between the continuous variables, we'll keep them for now
```

```python

# let's visualize the continuous variables w.r.t the target
for c in cnames:
    df[c].groupby(df['income']).mean().plot(kind='bar')
    plt.ylabel('average: '+c)
    plt.show()

    
#let's do ANOVA analysis to check the relationship of the continuous variables with the target 'income' categorical variable
from scipy import stats    
not_related=[]

for col in cnames:
    fStat, pVal = stats.f_oneway(df[col], df['income'])
    print(col,' P-value: ',pVal,' f-statistic: ',fStat)
    if pVal<0.05:
        print ('There is a relationship between col: ',col,' and the target variable')
    else:
        print ('No relationship between col: ',col,' and the target variable')
        not_related.append(col)
        
not_related       
# all show a relationship with the target 'income'
```

```python
# lets do a chi-square contingency test to check if there is a relationship between the categorical variables
print ('Categorical variables and target relations')

from scipy.stats import chi2_contingency
temp_cat = [c for c in cat_names if c!='income']
for i in temp_cat:
    print(i)
    chi2, p, dof, ex = chi2_contingency(pd.crosstab(df.income, df[i]))
    print(p)
    if p>0.05:
        print ('No relationship')
    else:
        print ('Relationship present')
    print ('\n')
    
# so all these categorical-variables have a relationship with the target 'income'    
```

```python
# let's perform two more feature selection techniques
# Let's perform a feature importance test
# this test gives you a score for each feature of your data, the higher score shows more importance/relevance 
# of that feature towards your target variable
# it uses a tree-based classifiers approach

from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.ensemble import ExtraTreesClassifier

X = df.iloc[:,:-1]
y = df['income']
importances = ExtraTreesClassifier().fit(X,y)
print (importances.feature_importances_)

# Let's look at the scores visually

values = pd.Series(importances.feature_importances_, index = df.columns[:-1])
values.nlargest(11).plot(kind='bar')
plt.figsize=(10,15)
plt.show()
# we shouldn't be removing any variable YET, because all of them more or less show some relationship to the target in ANOVA and
# chi-square analysis, so  we will decide after doing some further tests which ones to removing 
# also considering the importance-values received from this test.
```

```python
# source: https://towardsdatascience.com/feature-selection-using-regularisation-a3678b71e499
from sklearn.linear_model import Lasso, LogisticRegression # Lasso is L1 regularization
from sklearn.feature_selection import SelectFromModel

# fitting the feature selection model with the Logistic regression data
X = df.iloc[:,:-1]
sel_ = SelectFromModel(LogisticRegression(C=1, penalty='l1'))
sel_.fit(X, df['income'])


# Visualising features that were kept by the lasso regularisation
print (sel_.get_support())

#  So True is for the features that lasso thought is important (non-zero features) 
# while False is for the features whose weights were shrinked to zero and are not important according to Lasso.


selected_feat = X.columns[(sel_.get_support())] # gives the remaining features

print ('lasso-logistic features selected: ')
print (selected_feat)

from sklearn.ensemble import RandomForestClassifier
sel_ = SelectFromModel(RandomForestClassifier())

sel_.fit(X, df['income'])

# Visualising features that were kept by the lasso regularisation
print (sel_.get_support())
selected_feat = X.columns[(sel_.get_support())] # gives the remaining features
print ('Random-forest based features selected: ')
print (selected_feat)

# lasso-logistic feature selection model is showing like previous other tests that all the existing variables are necessary
# the Random-Forest feature selection method is shwoing that only ['age', 'fnlwgt', 'educational-num', 'relationship'] variables
# are important
# we'll create another dataframe with just ['age', 'fnlwgt', 'educational-num', 'relationship'] variables
# and develop some models over both dataframes to see which one gives better results, does the second dataframe actually reduce
# overfitting
```

```python
# creating another dataframe where continuous variables are not normalized, the original dataframe is normalized
# normalizing continuous data
df2 = df[['age', 'fnlwgt', 'educational-num', 'relationship','income']]    
df.head()

# lets begin now with the splitting of data into training and testing
from sklearn.cross_validation import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(df.iloc[:,:-1],df.iloc[:,-1]) # df split
X_train2,X_test2,Y_train2,Y_test2 = train_test_split(df2.iloc[:,:-1],df2.iloc[:,-1]) # df2 split


from sklearn.metrics import accuracy_score

def error_metrics(y_test,Y_preds):
    cm =pd.crosstab(y_test,Y_preds)
    TN = cm.iloc[0,0]
    FP = cm.iloc[0,1]
    FN = cm.iloc[1,0]
    TP = cm.iloc[1,1]
    print ('FNR: ',(FN/(FN+TP))*100)
    print ('accuracy: ',accuracy_score(y_test,Y_preds))
```

```python
from sklearn.linear_model import LogisticRegression

# original df
LR =  LogisticRegression().fit(X_train,Y_train)
LR_preds = LR.predict(X_test)
error_metrics(Y_test,LR_preds)

# df2
LR2 =  LogisticRegression().fit(X_train2,Y_train2)
LR_preds2 = LR2.predict(X_test2)
error_metrics(Y_test2,LR_preds2)
```

```python
from sklearn.ensemble import RandomForestClassifier

RF_model = RandomForestClassifier().fit(X_train,Y_train)
RF_preds = RF_model.predict(X_test)
error_metrics(Y_test,RF_preds)

RF_model2 = RandomForestClassifier().fit(X_train2,Y_train2)
RF_preds2 = RF_model2.predict(X_test2)
error_metrics(Y_test2,RF_preds2)
```

```python
# comparing the first few models applied over df and df2, there is higher FNR and lower accuracy in df2
# so we stopped applying models after the first three on df2 (concluding that random-forest feature selection over this dataset)
# we'll use df only for the deep-learning model
```

```python
#Define a variable for each type of feature
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

target = ["income"]

numeric_columns = cnames
categorical_columns = cat_names[:-1]
#Define a function that will intake the raw dataframe and the column name and return a one hot encoded DF
def create_ohe(df2, col):
    le = LabelEncoder()
    a = le.fit_transform(df2[col]).reshape(-1,1)
    ohe = OneHotEncoder(sparse=False)
    column_names = [col+ "_"+ str(i) for i in le.classes_]
    return(pd.DataFrame(ohe.fit_transform(a),columns =column_names))

#Since the above function converts the column, one at a time
#We create a loop to create the final dataset with all features
temp = df[numeric_columns]
for column in categorical_columns:
    temp_df = create_ohe(df,column)
    temp = pd.concat([temp,temp_df],axis=1)

print("Shape of Data:",temp.shape)
print("Distinct Datatypes:",temp.dtypes.unique())
temp.head()
```

```python
k=0
for i in cat_names[:-1]:
    k = k+len(df[i].unique())
    
k    
```

```python
# import the necessary modules for DNN
from keras.models import Sequential
from keras.layers import Dense, Activation

# define the DNN model
model = Sequential()
model.add(Dense(256, input_dim=105,activation = "relu")) # 105 input features, twice the number of neurons:256 in 1st hidden layer
#model.add(Dense(256,activation = "relu"))
model.add(Dense(128,activation = "relu"))
model.add(Dense(1,activation = "sigmoid")) # because the target is binary classification
model.compile(optimizer='rmsprop',loss='binary_crossentropy',metrics=['accuracy'])

# split the data for training and validation
train_data,test_data,train_targets,test_targets = train_test_split(temp,df['income'],test_size=0.3,random_state=42)
ptrain_data ,x_val,ptrain_targets,y_val = train_test_split(train_data,train_targets,test_size=0.2,random_state=42)#validation data
print (ptrain_data.shape ,ptrain_targets.shape)
print (x_val.shape,y_val.shape)
```

```python
history = model.fit(ptrain_data,ptrain_targets,batch_size=32,epochs=50,validation_data=(x_val, y_val))

preds = model.predict(test_data)
test_results = model.evaluate(test_data,test_targets)
for i in range(len(test_results)):
    print (model.metrics_names[i],': ',test_results[i])

preds = [round(i[0]) for i in preds]
error_metrics(test_targets,np.array(preds))

import matplotlib.pyplot as plt
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(1, len(loss) + 1)
plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

# loss :  0.415793634525527
# acc :  0.8334129529504379
# FNR:  43.07017543859649
# accuracy:  0.8334129529789122
# the validation error seen in the graph seems to be increasing with epochs 
```
